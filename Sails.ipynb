Pridedu nauja stulpeli

```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

Istraukiu nauja stulpeli ir sulyginu su pardavimais

```python
metai_pardavimai = df.groupby('month_year').sum()['sales'].reset_index()
print(metai_pardavimai)
```

```python
plt.plot(metai_pardavimai['month_year'], metai_pardavimai['sales'])
plt.xlabel('Mėnesis ir Metai')
plt.ylabel('Pardavimai')
plt.title('Pardavimai pagal Mėnesį ir Metus')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

Top 10 produktu

```python
produktai = df.groupby('product_name')['sales'].sum()
produktai = produktai.reset_index()
print(produktai)
```

```python
prod_parda= produktai.sort_values('sales', ascending=False)
```

```python
prod_parda[:10]
```

Daugiausiai parduotu produktu

```python
max_parduotu = df.groupby('product_name')['quantity'].sum().sort_values(ascending=False)
max_parduotu = max_parduotu.reset_index()
print(max_parduotu)
```

```python
max_parduotu[:10]
```
